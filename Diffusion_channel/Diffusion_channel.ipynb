```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Read both CSVs
df1 = pd.read_csv('3GPP.csv')
df2 = pd.read_csv('/content/Quadriga.csv')

# Create a figure with 1 row and 2 columns
fig, axes = plt.subplots(1, 2, figsize=(10, 4))  # (width, height)

# Plot for 3GPP
axes[0].plot(df1['SNR'], df1['nmse_dm'], marker='o', linestyle='-', color='b')
axes[0].set_yscale('log')
yticks1 = np.logspace(np.log10(df1['nmse_dm'].min()), np.log10(df1['nmse_dm'].max()), num=5)
axes[0].set_yticks(yticks1)
axes[0].set_yticklabels([f'10^{int(np.log10(val))}' for val in yticks1])
axes[0].set_xlabel('SNR (dB)')
axes[0].set_ylabel('Normalised MSE')
axes[0].set_title('MSE Performance - 3GPP')
axes[0].grid(True)

# Plot for QuaDRiGa
axes[1].plot(df2['SNR'], df2['nmse_dm'], marker='o', linestyle='--', color='r')
axes[1].set_yscale('log')
yticks2 = np.logspace(np.log10(df2['nmse_dm'].min()), np.log10(df2['nmse_dm'].max()), num=5)
axes[1].set_yticks(yticks2)
axes[1].set_yticklabels([f'10^{int(np.log10(val))}' for val in yticks2])
axes[1].set_xlabel('SNR (dB)')
axes[1].set_ylabel('Normalised MSE')
axes[1].set_title('MSE Performance - QuaDRiGa')
axes[1].grid(True)

# Adjust spacing between the plots
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Read both CSVs
df1 = pd.read_csv('3GPP.csv')
df2 = pd.read_csv('Quadriga.csv')

# Create the figure
plt.figure(figsize=(5,3))

# Plot the first curve
plt.plot(df1['SNR'], df1['nmse_dm'], marker='o', linestyle='-', color='b', label='3GPP')

# Plot the second curve
plt.plot(df2['SNR'], df2['nmse_dm'], marker='s', linestyle='--', color='r', label='QuaDRiGa')

# Log scale for y-axis
plt.yscale('log')

# Set dynamic yticks based on min/max of both curves
combined_min = min(df1['nmse_dm'].min(), df2['nmse_dm'].min())
combined_max = max(df1['nmse_dm'].max(), df2['nmse_dm'].max())
yticks = np.logspace(np.log10(combined_min), np.log10(combined_max), num=5)
plt.yticks(yticks, [f'10^{int(np.log10(val))}' for val in yticks])

# Labels, title, grid, legend
plt.xlabel('SNR (dB)')
plt.ylabel('Normalized MSE')
plt.title('MSE Performance - 3GPP vs QuaDRiGa')
plt.grid(True, which='both', linestyle='--', linewidth=0.5)
plt.legend()

# Show the plot
plt.tight_layout()
plt.show()
```

```python
csv_files = glob.glob('Plots/*.csv')
colors = plt.cm.get_cmap('tab20', len(csv_files))

plt.figure(figsize=(5,3))

for idx, file in enumerate(csv_files):
    df = pd.read_csv(file)
    plt.plot(df['t'], df['nmse_dm'], marker='o', markersize=3, linewidth=1,
             linestyle='-', color=colors(idx),
             label=os.path.basename(file))

plt.yscale('log')
plt.gca().invert_xaxis()
plt.xticks([75, 50, 25, 0])
plt.yticks([10**0, 10**-1, 10**-2], ['10^0', '10^-1', '10^-2'])

plt.xlabel('DM Step l')
plt.ylabel('Normalised MSE')
plt.title('MSE Performance over intermediate steps - 3GPP')
plt.grid(True, which='both', linestyle='--', linewidth=0.2)
plt.legend(fontsize='small', loc='best')
plt.tight_layout()
plt.show()
```
